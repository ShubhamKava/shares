```python
symbol = input('Enter company symbol name\n')
```

```python
duration = input('Enter how long data you want to work\n')
```

```python
df = yf.download(symbol , period = duration)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop('Adj Close', axis = 1, inplace=True)
```

```python
df['SMA_5'] = df['Close'].rolling(5).mean()
```

```python
df['SMA_20'] = df['Close'].rolling(20).mean()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.head(20)
```

```python
df.info()
```

```python
Action = []
```

```python
df['Action'] = np.where(df["SMA_5"] > df["SMA_20"],"S", "B" )
```

```python
df.head(100)
```

```python
df['Buy / Sales'] = np.where(df['Action'] == 'S', "Sales", "Buy")
```

```python
# Reset the index to move the date to a column
df = df.reset_index()
```

```python
df.head()
```

```python
for i in range(1, len(df)):
    if df['Action'][i] == df['Action'][i-1]:#  & df['Action'][i-2]
        df['Buy / Sales'][i]= 'Wait. . '
    elif df['Action'][i-1] == 'S':
        df['Buy / Sales'][i]= 'Buy'
    else:
        df['Buy / Sales'][i] = 'Sales'
        
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18,8),dpi=250)
plt.plot(df['Date'], df[['SMA_5','SMA_20', 'Close']], label=['SMA 5','SMA 20', 'Close Price'])
plt.xlabel('Date')
plt.ylabel('SMA Values')
plt.legend(loc = 'best')
plt.title('Simple Moving Average with several intervals and Closing price')
plt.grid()
plt.show();
```

```python
colors = ['yellow', 'green','red' ]
# label = df['Buy / Sales2'
plt.pie(x = df['Buy / Sales'].value_counts(normalize=True), colors=colors, labels=(['Wait . .', 'Buy','Sales']), autopct='%1.1f%%')
plt.legend()
plt.title("Pieplot for Buy / Sales Option");

print('Options Percentage: -')
print(df['Buy / Sales'].value_counts(normalize=True))
print('\n')

print('Count number of option: -')
print(df['Buy / Sales'].value_counts())
```
